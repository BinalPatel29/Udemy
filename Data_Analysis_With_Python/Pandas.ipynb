```python
## series

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
## dictionary

data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## create a dataframe from a dict of list

data={
    'Name':['Binal','Bhoomi','Riya'],
    'Age':[25,30,45],
    'city':['Banglore','New York','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## create a dataframe from a list of dict

data=[
    {'Name':'Binal','Age':32,'city':'Banglore'}, 
    {'Name':'Bhoomi','Age':33,'city':'New York'},
    {'Name':'Riya','Age':34,'city':'Florida'},
    {'Name':'Drashti','Age':35,'city':'Chikago'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
