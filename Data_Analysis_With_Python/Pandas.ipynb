```python
## renaming columns
df=df.rename(columns={'Sale_Date':'Sales Date'})
df=df.rename(columns={'Quantity_Sold':'New_Value'})
df.head()
```

```python
df['Value_New']=df['New_Value'].apply(lambda x:x*2)
df
```

```python
## data aggregating and grouping
df.head(5)
```

```python
grouped_mean=df.groupby('Product_Category')['New_Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product_Category','Region'])['New_Value'].sum()
grouped_sum
```

```python
df.groupby(['Product_Category','Region'])['New_Value'].mean()
```

```python
## aggregate multiple function
grouped_agg=df.groupby('Region')['New_Value'].agg(['mean','sum','count'])
grouped_agg
```

```python
## merging and joining dataframe
# create sample dataframe
df1=pd.DataFrame({'key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'Value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## merge dataframe on the 'key column'
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
