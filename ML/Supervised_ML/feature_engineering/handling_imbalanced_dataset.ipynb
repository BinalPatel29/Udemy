```python
## two tech:  1)up sampling   2)down sampling

import numpy as np
import pandas as pd
```

```python
## set the  random seed for reproductivity
np.random.seed(123)

## create a dataframe with the classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = n_samples-n_class_0 
```

```python
n_class_0,n_class_1
```

```python
## create dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1 ,size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1 ,size=n_class_0),
    'target':[0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1 ,size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1 ,size=n_class_1),
    'target':[1] * n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
df.tail()
```

```python
df['target'].value_counts()
```

```python
## upsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
!pip install scikit-learn
from sklearn.utils import resample
```

```python
df_minority_majority=resample(df_minority,replace=True,
         n_samples=len(df_majority),
         random_state=42
         )
```

```python
df_minority_majority.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_majority])
```

```python
df_upsampled['target'].value_counts()
```

```python
# down sampling
```

```python
## create dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1 ,size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1 ,size=n_class_0),
    'target':[0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1 ,size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1 ,size=n_class_1),
    'target':[1] * n_class_1
})

df=pd.concat([class_0,class_1]).reset_index(drop=True)

# check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_majority_downsample=resample(df_majority,replace=False,
         n_samples=len(df_minority),
         random_state=42
         )
```

```python
df_majority_downsample.shape
```

```python
df_downsampled=pd.concat([df_minority,df_majority_downsample])
```

```python
df_downsampled.target.value_counts()
```
