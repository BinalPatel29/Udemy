```python
## its simple way to create iterators
## use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory
```

```python
def square(n):
    for i in range(3):
        return i**2
```

```python
square(3)    ## only give the first element value
```

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):    ## yeild give all elements output
    print(i)
```

```python
a=square(3)
a
print(next(a))
print(next(a))
print(next(a))
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_generator()
gen
```

```python
next(gen)    ## after 3 times of cell run (1,2,3,error)
```

```python
for val in gen:
    print(val)
```

```python
## practical: reading large file --> generators use

def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
```

```python
file_path="large_file.txt"

for line in read_large_file(file_path):
    print(line.strip())
```
