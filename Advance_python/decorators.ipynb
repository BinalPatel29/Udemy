```python
# decorators

def main_welcome(func):
    def sub_welcome_method():
        print("welcome to the advance python")
        func()
        print("Please learn there concepts properly")
    return sub_welcome_method()
```

```python
def course_intro():
    print("this is an advanced python")

course_intro()
```

```python
main_welcome(course_intro)
```

```python
@main_welcome
def course_intro():
     print("this is an advanced python")
```

```python
# decorator

def my_decorator(func):
    def wrapper():
        print("something is happening before the function is called.")
        func()
        print("something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello")
```

```python
say_hello()
```

```python
## decorator 
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```
