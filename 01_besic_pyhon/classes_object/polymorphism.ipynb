```python
## allows objects of different classes to be treates as objects of a common superclass
## provides way to perform a single action in different forms
```

```python
## Method Overriding (same method)
class Animal:
    def speak(self):
        return "Sound of the animal"
    
## derived class 1
class Dog(Animal):
    def speak(self):
        return "woof!"
    
## derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow!"
    
## function that demonstrats polymorphism
def animal_speak(animal):
    print(animal.speak())
    
dog=Dog()
print(dog.speak())
cat=Cat()
print(cat.speak())
animal_speak(dog)
```

```python
## polymorphism with function and methods
## base class
class Shape:
    def area(self):
        return "The area of the figure"
    
## derived class 1
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height  
    def area(self):
        return self.width*self.height
    
## derived class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.14*self.radius*self.radius
    
## function that demonstrate polymorphism
def print_area(shape):
    print(f"area is {shape.area()}")

rectangle=Rectangle(4,5)
circle=Circle(3)

print_area(rectangle)
print_area(circle)
        
```

```python
## polymorphism with abstract base class
## abstract base class are used to define common methods for a group of related objects
from abc import ABC,abstractmethod

## define an object class
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

## derived class 1
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
## derived class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"
    
##function demontrates polymorphism
def start_vehicle(vehicle):
    print(vehicle.start_engine())

## create objects
car=Car()
motorcycle=Motorcycle()

start_vehicle(car)
```
