```python
## data serialization
import json
data={'name':'binal','age':22}

json_str=json.dumps(data)     ## convert into json format  (string)
print(json_str)
print(type(json_str))

parsed_data=json.loads(json_str)    ## for get return back dict
print(parsed_data)
print(type(parsed_data))
```

```python
## csv
import csv
with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['bianl',22])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row) 
```
