1) Nomial/OHE encoding--- one hot encoding

```python
# covert categorical data into numeric data
# use only minimum(small) data

import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# create a simple dataframe
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
# create an instance of OHE
encoder=OneHotEncoder()
```

```python
# perform fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
# ordinal encoder
# convert categorical data into numeric data
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df= pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df
```

```python
## craete an instance of OrdinalEncoder and then fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

3.Target Guided Ordinal Encoding

```python
# used to encode categorical variables based on the relationship with the target value
# use mean and median
```

```python
import pandas as pd

# create a sample dataframe with a categorical variable and a target variable
df= pd.DataFrame({
    'city':['new york','london','paris','tokyo','new york','paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df[['city','city_encoded']]
```
