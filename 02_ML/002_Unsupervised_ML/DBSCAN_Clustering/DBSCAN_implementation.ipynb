```python
## DBSCAN== Density-Based Spatial Clustering of Applications with Noise

## DBSCAN 3 types ke points define karta hai:
# Core points → jinke aas-paas enough neighbors ho
# Border points → cluster ke edge par
# Noise points (outliers) → kisi cluster me fit nahi hote

## DBSCAN 2 parameters use karta hai:
# eps (ε)
# → kitni distance tak points ko “neighbors” mana jaye
# min_samples
# → minimum kitne points hone chahiye ek dense area banane ke liye
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=250,noise=0.05)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
## feature scalling(standard scalling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.3)
```

```python
dbscan.fit(x_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=dbscan.labels_)
```
