```python
from flask import Flask, request

# WSGI Application
app = Flask(__name__)

@app.route('/')
def welcome():
    return "Welcome to this flask course!"

@app.route('/index')
def index():
    return "Welcome to the index page!"

if __name__ == '__main__':
    app.run(debug=True, use_reloader=False, port=5001)
```
